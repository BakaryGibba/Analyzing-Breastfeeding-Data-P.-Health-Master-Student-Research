### **1. Chi-square Test**
Test if knowledge levels vary by caregiver sex.

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_relevant['Caregiver Sex_2. Female'], df_relevant['High Knowledge'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}, P-value: {p}")
```

### **2. Independent t-test**
Compare knowledge scores between mothers and other caregivers.



```python
# Example: Compare knowledge scores between mothers and other caregivers
# Since 'Relationship to Child' is not one-hot encoded, we'll filter directly
mothers = df_relevant[df_relevant['Relationship to Child'] == '1. Mother']['Total Knowledge Score']
others = df_relevant[df_relevant['Relationship to Child'] != '1. Mother']['Total Knowledge Score']

from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(mothers, others)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

### **3. Logistic Regression**
Predict high knowledge levels based on caregiver characteristics.



```python
import statsmodels.api as sm
import pandas as pd
import numpy as np

# ... (your existing code to create df_relevant) ...

# One-hot encode 'Education Level' and 'Caregiver Sex' (if needed)
columns_to_encode = ['Education Level', 'Caregiver Sex']
for col in columns_to_encode:
    if col in df_relevant.columns and df_relevant[col].dtype == 'object':
        df_relevant = pd.get_dummies(df_relevant, columns=[col], prefix=[col], drop_first=False)

# Define independent and dependent variables
X = df_relevant[['Caregiver Age', 'Number of Children', 'Caregiver Sex_2. Female',
                 'Education Level_2. Secondary', 'Education Level_3. Higher']]

# Ensure all columns are numeric
for col in X.columns:
    if X[col].dtype == 'object':
        try:
            X[col] = pd.to_numeric(X[col], errors='raise')
        except ValueError:
            # Impute non-numeric values with a suitable strategy (e.g., median)
            # You can handle different column types (e.g., numbers as strings, strings that can be converted to numbers)
            X[col] = pd.to_numeric(X[col], errors='coerce')  # Convert errors to NaN
            X[col].fillna(X[col].median(), inplace=True) # Fill NaN with median
    # Explicitly convert to float to avoid potential dtype issues with Statsmodels
    X[col] = X[col].astype(float)


X = sm.add_constant(X)  # Add a constant for the intercept
y = df_relevant['High Knowledge']

# Fit the logistic regression model
model = sm.Logit(y, X)
result = model.fit()
print(result.summary())
```
